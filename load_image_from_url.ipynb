## Load Image From URL

```python
import urllib.request
from PIL import Image
import pandas as pd 
import matplotlib.pyplot as plt 

cols = ['ImageName','FileDir','CollectDate','Instrument']
filepath = 'WebbDemo.csv'

df = pd.read_csv(filepath, header=0, usecols=cols,skip_blank_lines=True,encoding='utf-8')#'latin1'
image_names = df['ImageName'].tolist()
image_urls = df['FileDir'].tolist()

index = 0
image_url = image_urls[index] 
image_name = image_names[index]
urllib.request.urlretrieve(image_url, 'image1')

img = Image.open('image1').convert('L')
plt.imshow(img, cmap='gray')
```
